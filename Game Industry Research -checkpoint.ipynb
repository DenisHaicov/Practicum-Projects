### Take a look at how user and professional reviews affect sales for one popular platform (you choose). Build a scatter plot and calculate the correlation between reviews and sales. Draw conclusions.

```python
def dif_platform(name_platform):
    platform = relevant_data[relevant_data['platform']==name_platform]
    fig,ax = plt.subplots(1,2,figsize=(15,10))
    sns.scatterplot(x='user_score', y='total_sales',data = platform, ax=ax[0], facecolor='red')
    sns.scatterplot(x='critic_score', y='total_sales',data = platform, ax=ax[1], facecolor='blue')
    fig.suptitle(name_platform, fontsize=15)
    ax[0].set(xlabel='user_score')
    ax[0].set(ylabel='total_sales')
    ax[1].set(xlabel='critic_score')
    ax[1].set(ylabel='total_sales')
    plt.show()
    
    user_corr = platform['user_score'].corr(platform['total_sales'])
    critic_corr = platform['critic_score'].corr(platform['total_sales'])
    
    if critic_corr >= 0.7:
        print('Correlation between critic score and platform', name_platform, ': Strong', critic_corr)
    elif 0.7 >= critic_corr >= 0.3:
        print('Correlation between critic score and platform', name_platform, ': Average', critic_corr)
    elif 0.3 >= critic_corr > 0:
        print('Correlation between critic score and platform', name_platform, ': Weeak', critic_corr)
        
    if user_corr >= 0.7:
        print('Correlation between user score and platform', name_platform, ': Strong', user_corr)
    elif 0.7 >= user_corr >= 0.3:
        print('Correlation between user score and platform', name_platform, ': Average', user_corr)
    elif 0.3 >= user_corr >= 0:
        print('Correlation between user score and platform', name_platform, ': Weak', user_corr)

        print('\n')
```

# Test the following hypotheses:

### Average user ratings of the Xbox One and PC platforms are the same.

```python
xbox1 = relevant_data[(relevant_data['platform']=='XOne')&(relevant_data['user_score'].notna())]['user_score']
```

```python
xbox1
```

```python
pc = relevant_data[(relevant_data['platform']=='PC')&(relevant_data['user_score'].notna())]['user_score']
```

```python
pc
```

```python
from scipy.stats import levene
stat, p = levene(xbox1, pc)
p<0.05
```

```python
from scipy import stats
stats.ttest_ind(xbox1, pc, equal_var=False)
```

```python
stats.ttest_ind(xbox1, pc).pvalue<0.05
```

### Average user ratings for the Action and Sports genres are different.

```python
action = relevant_data[(relevant_data['genre']=='Action')&(relevant_data['user_score'].notna())]['user_score']
```

```python
action
```

```python
sports = relevant_data[(relevant_data['genre']=='Sports')&(relevant_data['user_score'].notna())]['user_score']
```

```python
sports
```

```python
stat, p = levene(action, sports)
p<0.05
```

```python
stats.ttest_ind(action, sports, equal_var=False)
```

```python
stats.ttest_ind(action, sports).pvalue<0.05
```

My null hypothesis was that the average user ratings of Action and Sports genre will be the same.
The alternative hypotheses was that the average wone be the same.

After testing my hypotheses we can notice that the pvalue is greater then 0.05 alpa as we mentioned before(less then 5%) and the hypotheses was right.
